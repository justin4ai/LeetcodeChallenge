### (M) 54. Spiral Matrix

43 / 18

```python
matrix = [[1,2,3],[4,5,6],[7,8,9]]

class Solution(object):
    def spiralOrder(self, matrix):
        ans = []
        row, col = 0, 0
        i, j = 0, 1

        for s in range(len(matrix) * len(matrix[0])):
          
          #print(f"row and col are each {row} {col}")
          ans.append(matrix[row][col])
          matrix[row][col] = 'X'


          if ((row + i) < 0) or ((col + j) < 0) or ((row + i) > len(matrix) - 1) or ((col + j) > len(matrix[0]) - 1):
            i, j = j, -i

          else:
            if (matrix[row + i][col + j] == 'X'):
              i, j = j, -i  


          row += i
          col += j
        return ans 

print(Solution().spiralOrder(matrix))
        
```

```python
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(4)
head.next.next.next.next = ListNode(5)
```

Approach 1: Make a cycle and unlink the proper part later - 16 - 55

```python
k = 4

class Solution(object):
    def rotateRight(self, head, k):

        if head == None:
          return None

        cursor = head
        myStack = []

        while True:
          myStack.append(cursor)

          if cursor.next == None:
            cursor.next = head
            break

          cursor = cursor.next

        depth = len(myStack)
        newStack =  myStack[-(k % depth):] +myStack[:-(k % depth)]

        newStack[-1].next = None

        return newStack[0]

print(Solution().rotateRight(head, k))
```

### (E) 171. Excel Sheet Column Number

Recursive calling - 28 / 27

```python
columnTitle = 'FXSHRXW'

class Solution(object):
    def titleToNumber(self, columnTitle):
        
        def func(columnTitle):

            if len(columnTitle) == 1:
              return ord(columnTitle) - 64

            return func(columnTitle[-1]) + 26 * func(columnTitle[:-1])
        
        return func(columnTitle)

print(Solution().titleToNumber(columnTitle))
```

Iterative - 18 / 80

```python
class Solution(object):
    def titleToNumber(self, columnTitle):

      strLen = len(columnTitle)
      sum = 0

      for idx in range(strLen):
        sum += (ord(columnTitle[strLen - idx - 1]) - 64) * (26**idx)

      return sum
```

### (E) 168. Excel Sheet Column Title â˜…

Iterative method - 56 / 36

```python
columnNumber = 2147483647

class Solution(object):
    def convertToTitle(self, columnNumber):
        pre = 1
        N = 1

        while True:
          if pre <= columnNumber < 26 * pre + 1:
            break

          N += 1
          pre = 26 * pre + 1

        string = ''
        for n in range(N, 0, -1): # from N to 1 : N times in total
          
          if (columnNumber % (26 ** (n - 1))) == 0:
            if n == 1:
              asc = columnNumber
            else:
              asc = (columnNumber-1) // 26 ** (n - 1)
          else:  
            asc = columnNumber // 26 ** (n - 1)

          string += chr(asc + 64)
          columnNumber -= (asc * 26 ** (n - 1))


        return string
        
print(Solution().convertToTitle(columnNumber))
```
