### (M) 98. Validate Binary Search Tree ★

image.png

```python
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

root = TreeNode(5)
root.left = TreeNode(1)
root.right = TreeNode(4)
root.right.left = TreeNode(3)
root.right.right = TreeNode(6)
```

Recursive way - 60 / 84 <br><br>
**※ I gave up on this problem exactly two weeks ago, but now easily solved this!!!**

```python
class Solution(object):
    def isValidBST(self, root):
        self.ans = True

        def dfs(CurNode, range):
          if range[0] < CurNode.val < range[1]:
            if CurNode.left != None:
              dfs(CurNode.left, [range[0], min(range[1], CurNode.val)])
            if CurNode.right != None:
              dfs(CurNode.right, [max(range[0], CurNode.val), range[1]])

          else:
            self.ans = False
            return

        dfs(root, [-2**31 - 1, 2**31 + 2])
        return self.ans
        
Solution().isValidBST(root)
```

### (E) 108. Convert Sorted Array to Binary Search Tree

Recursive DFS let's go - 30 / 74<br>
It was the first time to handle BST with code but I can feel I'm on level up

```python
nums = [-10,-3,0,5,9]
class TreeNode(object):
   def __init__(self, val=0, left=None, right=None):
       self.val = val
       self.left = left
       self.right = right

class Solution(object):
    def sortedArrayToBST(self, nums):
        def med(arr):
          return (len(arr) - 1) // 2
        
        root = TreeNode(nums[med(nums)])
        leftArr, rightArr = nums[:med(nums)], nums[med(nums) + 1:]

        def dfs(curNode, leftArr, rightArr):
          if leftArr != []:
            med_left = med(leftArr)
            curNode.left = TreeNode(leftArr[med_left])
            dfs(curNode.left, leftArr[:med_left], leftArr[med_left + 1:])

          if rightArr != []:
            med_right = med(rightArr)
            curNode.right = TreeNode(rightArr[med_right])
            dfs(curNode.right, rightArr[:med_right], rightArr[med_right + 1:])

        dfs(root, leftArr, rightArr)
        return root

Solution().sortedArrayToBST(nums)        
```
